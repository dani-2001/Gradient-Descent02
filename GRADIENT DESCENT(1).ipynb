```python
import numpy as np

class GDRegressor:
    def __init__(self, learning_rate=0.01, n_iterations=100):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.theta = None  # Model parameters (weights)

    def fit(self, X, y):
        X = np.array(X)
        y = np.array(y)
        
        m, n = X.shape  # m = number of samples, n = number of features
        self.theta = np.zeros(n)  # Initialize parameters (weights)

        for _ in range(self.n_iterations):
            gradients = (1 / m) * X.T.dot(X.dot(self.theta) - y)
            self.theta -= self.learning_rate * gradients

    def predict(self, X):
        return X.dot(self.theta)
```

```python
# Sample data (replace this with your actual dataset)
x = np.array([[1, 2], [2, 3], [3, 4], [4, 5]])
y = np.array([5, 7, 9, 11])

# Instantiate the model with learning rate and iterations
lgd = GDRegressor(learning_rate=0.01, n_iterations=100)

# Fit the model to your data
lgd.fit(x, y)

# Make predictions
x_test = np.array([[5, 6], [6, 7]])
predictions = lgd.predict(x_test)

print(predictions)
```
